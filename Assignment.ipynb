```python
def generate_students():
    
    students = [
        {'id':'DBU1500005', 'name': 'Tesfaye', 'dept': 'DS', 'cgpa': 3.34},
        {'id':'DBU1500009', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
        {'id':'DBU1500003', 'name': 'Girma', 'dept': 'DS', 'cgpa': 2.74},
        {'id':'DBU1500004', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.10},
        {'id':'DBU1500001', 'name': 'Mohammed', 'dept': 'SE', 'cgpa': 3.75},
        {'id':'DBU1500006', 'name': 'Helen', 'dept': 'IT', 'cgpa': 3.50},
        {'id':'DBU1500007', 'name': 'Worku', 'dept': 'CE', 'cgpa': 2.95},
        {'id':'DBU1500010', 'name': 'Elias', 'dept': 'CS', 'cgpa': 3.20},
        {'id':'DBU1500002', 'name': 'Demissie', 'dept': 'DS', 'cgpa': 3.90},
        {'id':'DBU1500008', 'name': 'Yared', 'dept': 'IT', 'cgpa': 3.05}
    ]
    return students

 
def bubble_sort_by_id(students):
    n = len(students)
    sorted_students = students.copy()
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if sorted_students[j]['id'] > sorted_students[j+1]['id']:
                sorted_students[j], sorted_students[j+1] = sorted_students[j+1], sorted_students[j]
                swapped = True
        if not swapped:
            break
    return sorted_students

def insertion_sort_by_name(students):
    sorted_students = students.copy()
    for i in range(1, len(sorted_students)):
        key = sorted_students[i]
        j = i-1
        while j >=0 and sorted_students[j]['name'] > key['name']:
            sorted_students[j+1] = sorted_students[j]
            j -= 1
        sorted_students[j+1] = key
    return sorted_students

def selection_sort_by_cgpa(students):
    sorted_students = students.copy()
    for i in range(len(sorted_students)):
        max_idx = i
        for j in range(i+1, len(sorted_students)):
            if sorted_students[j]['cgpa'] > sorted_students[max_idx]['cgpa']:
                max_idx = j
        sorted_students[i], sorted_students[max_idx] = sorted_students[max_idx], sorted_students[i]
    return sorted_students
 
students = generate_students()

print("Original Student List:")
print("-" * 60)
for student in students:
    print(f"ID: {student['id']}, Name: {student['name']:20}, Dept: {student['dept']}, CGPA: {student['cgpa']:.2f}")

 
print("\nStudents Sorted by ID (Bubble Sort):")
print("-" * 60)
sorted_by_id = bubble_sort_by_id(students)
for student in sorted_by_id:
    print(f"ID: {student['id']}, Name: {student['name']}")

print("\nStudents Sorted by Name (Insertion Sort):")
print("-" * 60)
sorted_by_name = insertion_sort_by_name(students)
for student in sorted_by_name:
    print(f"Name: {student['name']:20}, ID: {student['id']}")

print("\nStudents Sorted by CGPA (Selection Sort - Descending):")
print("-" * 60)
sorted_by_cgpa = selection_sort_by_cgpa(students)
for student in sorted_by_cgpa:
    print(f"CGPA: {student['cgpa']:.2f}, Name: {student['name']:20}, ID: {student['id']}")
```

<h2>1.Bubble Sort (Sorting by ID)</h2>

Performance

      1.Time Complexity:<br>
       ✓Best Case: O(n)<br>
       ✓ Average/Worst Case: O(n²)<br>
       ✓ Space Complexity: O(1)<br>


Advantages<br>

Simple and easy to implement.<br>
Can be optimized to detect if the list is already sorted and exit early.<br>
Helpful for learning basic sorting concepts, especially comparisons and swapping.<br>

Disadvantages<br>

Very inefficient for large datasets due to many unnecessary comparisons.<br>
Poor performance even if the list is nearly sorted (unless optimized).<br>
Not cache-friendly—frequent swaps can slow performance on large lists.<br>

 <h2>2. Insertion Sort (Sorting Alphabetically by Name)</h2>
 
 Performance
    Time Complexity:<br>
         ✓ Best Case: O(n) (when the list is already sorted)<br>
         ✓ Average/Worst Case: O(n²)<br>
         ✓ Space Complexity: O(1)<br>

Advantages<br>

Stable sort: preserves the original order of students with the same name.<br>
Efficient for small or nearly sorted datasets.<br>
Suitable for real-time or streaming systems where data is added one record at a time.<br>
Allows partial sorting, which can be useful in dynamic lists.<br>

Disadvantages<br>

Performance drops significantly on reverse-ordered data.<br>
Still inefficient for large datasets due to quadratic time complexity.<br>

<h2>3.Selection Sort (Sorting by CGPA in Descending Order)</h2>

Performance<br>

Time Complexity: Always O(n²) (regardless of initial list order)<br>
Space Complexity: O(1)<br>

Advantages<br>

Very simple and predictable behavior.<br>
Performs the minimum number of swaps, which is useful when write operations are expensive (e.g., in memory-constrained systems).<br>
Easy to implement and understand.<br>

Disadvantages<br>

Not a stable sort: students with the same CGPA may end up in random order.<br>
Not efficient for large datasets due to consistently high comparison count.<br>
Does not take advantage of partially sorted data."""<br>
